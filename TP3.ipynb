```python
from statistics import mean
```

```python

def kMeans(img, c1, c2, c3, c4):
  img1 = img - c1
  img2 = img - c2
  img3 = img - c3
  img4 = img - c4
  arr = np.array([img1, img2, img3, img4])
  min_ind = np.argmin(arr, axis = 0)
  mask_1 = (min_ind == 0)
  mean_1 = mean((img*mask_1).flatten())
  mask_2 = (min_ind == 1)
  mean_2 = mean((img*mask_2).flatten())
  mask_3 = (min_ind == 2)
  mean_3 = mean((img*mask_3).flatten())
  mask_4 = (min_ind == 3)
  mean_4 = mean((img*mask_4).flatten())
  return min_ind, mean_1, mean_2, mean_3, mean_4

```

```python
seg, c1, c2, c3, c4 = kMeans(img_arr,c1, c2, c3, c4)
seg.shape
```
